```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
# Opening JSON files
df0 = json.load(open('/content/drive/MyDrive/Data Spotify/MyData/StreamingHistory0.json', 'r'))
df1 = json.load(open('/content/drive/MyDrive/Data Spotify/MyData/StreamingHistory1.json', 'r'))
df2 = json.load(open('/content/drive/MyDrive/Data Spotify/MyData/StreamingHistory2.json', 'r'))
df3 = json.load(open('/content/drive/MyDrive/Data Spotify/MyData/StreamingHistory3.json', 'r'))
df4 = json.load(open('/content/drive/MyDrive/Data Spotify/MyData/StreamingHistory4.json', 'r'))
```

```python
# Transform into Data Frame
df0 = pd.DataFrame(df0)
df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)
df3 = pd.DataFrame(df3)
df4 = pd.DataFrame(df4)
```

```python
# Merge Data Frames
df = pd.concat([df0, df1, df2, df3, df4])
df.head()
```

```python
df.shape
```

```python
# Count the most played artist
y = df["artistName"].value_counts().head(10)
```

```python
# Plot
fig = plt.subplots(1,1, figsize=(12,6))
y.plot.bar(color=mcolors.TABLEAU_COLORS)
plt.xlabel("Artist Name")
plt.ylabel("Times listened")
plt.title("Most played artist")
plt.show()
```

```python
# Count the most played song of fav artist
y1 = df[df['artistName'] == "Kevin Kaarl"].value_counts("trackName").head(10)
```

```python
# Plot
fig = plt.subplots(1,1, figsize=(12,6))
y1.plot.bar(color=mcolors.TABLEAU_COLORS)
plt.xlabel("Song Name")
plt.ylabel("Times listened")
plt.title("Most played son by my Top artist")
plt.show()
```

```python
# Count the most played track name
y2 = df["trackName"].value_counts().head(10)
```

```python
# Plot
fig = plt.subplots(1,1, figsize=(12,6))
y2.plot.bar(color=mcolors.TABLEAU_COLORS)
plt.xlabel("Song Name")
plt.ylabel("Times listened")
plt.title("Most played Song")
plt.show()
```
